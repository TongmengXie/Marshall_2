```python
import re
def clean(text):
    pattern = r'<p>(.*?)<\/p>'
    
    # Find all matches
    matches = re.findall(pattern, text)
    
    # Join the matches to form the full description
    description = ' '.join(matches)
    return description
patent.Description = patent.Description.apply(clean)
```
